```python
import os

# root path
parent_folder_path = '../Dataset/2017-10-01-juliet-test-suite-for-c-cplusplus-v1-3/C/testcases'

# list child folders
list_child_folders = []

# loop root folder
for item in os.listdir(parent_folder_path):
    # create child folder path
    item_path = os.path.join(parent_folder_path, item)
    
    # check is folder(dir) exists
    if os.path.isdir(item_path):
        print(f"Found folder: {item_path}")
        list_child_folders.append(item_path)
        # child folder
        for sub_item in os.listdir(item_path):
            sub_item_path = os.path.join(item_path, sub_item)
            if os.path.isfile(sub_item_path):
                print(f"    File in folder: {sub_item}")
```

```python
# delete comments 
import re
def remove_comments(code):
    # reguler expression delete comment
    # delete (//...)
    code = re.sub(r'//.*', '', code)
    
    # deltete (/*...*/)
    code = re.sub(r'/\*.*?\*/', '', code, flags=re.DOTALL)
        
    return code
```

```python
# remove ifdef endif
def remove_ifdef(code):
    pos = code.find("#endif", 0)
    while pos != -1:
        # code.
    code.replace('\n\n', '\n')
    return code
```
